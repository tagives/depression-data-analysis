```python
import matplotlib.pyplot as plt  # import the seaborn package with matplotlib.pyplot.
import seaborn as sns 
```

```python
import numpy as np  # import the other necessary packages for the analysis.
import pandas as pd
```

```python
depression_data_excel = pd.read_excel('finalcleanedd.xlsx') #read the cleaned dataset. 
print(depression_data_excel) 
```

```python
options = ['Female']
female_data = depression_data_excel[depression_data_excel['Sex'].isin(options)] # print the dataframe which shows the female's informations.
print('\nFemale Dataframe :\n',
      female_data)
```

```python
plt.figure(figsize = (13,13))  # dimensions of the rectangle 
sns.boxplot(data = female_data, x='Depressed', y='Living Expenses' , medianprops = {'Color': 'black', 'ls' : '-' , 'lw': 3}) # create a boxplot showing the x axis is depression y axis is living expenses, and show the median line.
plt.title('The Relationship between Depression and Living Expenses on Women', fontsize = 15, fontweight = 'bold') # title of boxplot
plt.xlabel('Depression') # x axis showing the depression
plt.ylabel('Living Expenses') # y axis living expenses
plt.show() #show the boxplot 
```

```python
plt.figure(figsize = (13,13)) # set the figure size.
sns.boxplot(data = depression_data_excel, x='Depressed', y='Lasting Investment', medianprops = {'color': 'black', 'ls':'-', 'lw':3}) # create a boxplot showing the x is depression y is lasting investment, and show the median line.
plt.title('The Relationship between Depression and Lasting Investment', fontsize = 15, fontweight = 'bold') # title of the plot.
plt.xlabel('Depression') # x axis showing the depression.
plt.ylabel('Lasting Investment') # y axis showing the lasting investment.
plt.show() # show the plot.
```
