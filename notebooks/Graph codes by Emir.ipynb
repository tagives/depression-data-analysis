```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
```

```python
dep = pd.read_excel("finalcleanedd.xlsx")
dep.head()
```

```python
plt.figure(figsize = (10,10)) # set the figure size 
sns.lmplot(data=dep, x="Gained Asset", y="Living Expenses", hue="Depressed",height = 10)
plt.title('The Relationship between Gained Asset and Living Expense by Depression Status',fontsize = 15, fontweight = "bold")

```
